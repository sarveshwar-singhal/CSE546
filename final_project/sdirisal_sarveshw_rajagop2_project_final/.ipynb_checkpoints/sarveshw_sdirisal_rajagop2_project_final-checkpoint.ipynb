```python
# Importing the normal Libraries
import numpy as np
import pandas as pd
from google.colab import output
import matplotlib.pyplot as plt

# Sklearn libraries
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.metrics import accuracy_score

# Tensorflow Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, regularizers
import keras.backend as k

# Torch Libraries
import torch
```

# Data Gathering and analyzing Observations

```python
class Data_collection:
  def __init__(self,path) -> None:
    try:
      self.data = pd.read_csv(path)
    except:
      self.data = pd.read_excel(path)
    self.Observation_data_frame = pd.DataFrame([])
    print("Current Execution :: Reading Data from File.")
    self.Creating_the_observations()

  def finding_the_difference(self, column_data_1, column_data_2 = None, find_change_rate = 0):
    if find_change_rate == 0:
      difference_1 = []
      for i in range(len(column_data_1)):
        present_volume = column_data_1[i]
        try:
          next_day_volume = column_data_1[i+1]
        except:
          next_day_volume = column_data_1[i]

        rate_change = (next_day_volume - present_volume)/ present_volume

        difference_1.append(rate_change)
        difference = np.array(difference_1)

    if find_change_rate == 1:
      difference = column_data_1 - column_data_2
    return difference
  def Normalization(self,column_data):
    max_value = np.max(column_data)
    min_value = np.min(column_data)
    range = max_value - min_value
    return ((column_data - min_value)/range)
  
  def Assigning_the_target_function(self,dataframe):
    Target_array = []
    for i in range(0,len(dataframe)):
      target = None
      if dataframe['Observation 1'][i] > 0 and dataframe['Observation 4'][i] >0 :
        target = 'BUY'
      elif dataframe['Observation 1'][i] < 0 and dataframe['Observation 4'][i] > 0:
        target = 'HOLD'
      else:
        target = 'SELL'
      Target_array.append(target)
    return np.array(Target_array)


  def Creating_the_observations(self):
    # Observation 1 = Open - close
    self.Observation_data_frame['Observation 1'] = self.finding_the_difference(self.data['Close'], self.data['Open'],1)

    # Observation 2 = High - Low
    self.Observation_data_frame['Observation 2'] = self.finding_the_difference(self.data['High'], self.data['Low'],1)

    # Observation 3 = Low - Close
    self.Observation_data_frame['Observation 3'] = self.finding_the_difference(self.data['Low'], self.data['Close'],1)

    # Observation 4 = Rate of the Volume Change (The increase in rate represents the strength)
    self.Observation_data_frame['Observation 4'] = self.finding_the_difference(self.data['Open'])

    print("\nCurrent Execution :: Gathered Observations from the Data.")

    self.Normalized_data = self.Observation_data_frame.apply(self.Normalization)

    print("\nCurrent Execution :: Normalizing the Observations.")

    self.Target_array = self.Assigning_the_target_function(self.Observation_data_frame)

    print("\nCurrent Execution :: Creating the Target values according the Assumptions.")

    self.Label_encoded = LabelEncoder().fit_transform(self.Target_array)
    self.One_hot_encoded = pd.get_dummies(self.Target_array)

    print("\nCurrent Execution :: Converting to the One Hot and Label Encoding.")
  
Object_DC = Data_collection('/content/microsoft_all.xlsx')
```

# Creating the Model of the Actor Critic

```python
class ACTOR_CRITIC:
  def __init__(self,object_) -> None:
    self.DC = object_
    self.Target = self.DC.Target_array
    self.Rewards_metric = {0: -7, 1:10, 2:5} # Reward function of the Model
    self.gamma_value =0.99                   # gamma Value used in the TD error
    self.input_dimension_advantage =5        # Input Dimension of the Actor Network
    self.input_dimension =4                  # The Input Dimension of the Critic and Policy Network
    self.rewards = np.array([])
    self.holds = 0
    self.no_of_days = 0
    self.income = 0

    self.Start_data = 100  # The training data starts form the row 100
    self.End_data =200     # The initial data end row
    self.data_size = 100    # The data divided into batches of the size 20
    self.Actor  = []       # Store the average loss of the Actor network while training
    self.Critic = []       # Store the average loss of the Critic network while training
    self.Policy = []       # Store the average loss of the Policy network while training
    self.Action_took = 5   
    self.track_income = [] # Tracking the income
    self.stocks = 0        # Show the number of stocks the customer holds
    self.BUY =  1
    self.SELL = 0
    self.HOLD = 0

    # Actor Model
    self.Actor_model = models.Sequential([layers.InputLayer(self.input_dimension_advantage),
                                     layers.Dense(32, activation='relu',name="Hidden1"),
                                     layers.Dense(3, activation=keras.activations.softmax,name='Output_Layer')
                                     ])
    # Critic Model
    self.Critic_model = models.Sequential([layers.InputLayer(self.input_dimension),
                                     layers.Dense(32, activation='relu',name="Hidden1"),
                                     layers.Dense(1, activation='relu',name='Output_Layer')
                                     ])
    # Policy Model
    self.Policy_model = models.Sequential([layers.InputLayer(4,name= "Input"),
                                     layers.Dense(32, activation='relu',name="Hidden1"),
                                     layers.Dense(3, activation='relu',name='Output_Layer')
                                     ])
    self.Critic_model.compile(loss = keras.losses.huber, optimizer = keras.optimizers.SGD(learning_rate = 0.1))
    self.Policy_model.compile(loss = keras.losses.sparse_categorical_crossentropy, optimizer = keras.optimizers.Adam(learning_rate = 0.8))
    self.Actor_model.compile(loss = keras.losses.sparse_categorical_crossentropy, optimizer = keras.optimizers.Adam(learning_rate = 0.5))

    (self.X_train, self.X_test, self.Y_train, self.Y_test) = train_test_split(np.array(self.DC.Observation_data_frame),np.array(self.DC.Label_encoded), random_state=10)
  
  # Function 1 => Return the discounted rewards values of the rewards which given as the input.

  def discounted_rewards(self,reward_list):   #### Function 1
    '''Calculating the cummulative reward list.'''
    Cummulative_reward_list_timestep =[] # Having the Cummulative rewards for ech time step
    discount_factor = 0.9
    j = 0
    for i in range(len(reward_list)-1,-1,-1):
      Cummulative_reward_list_timestep.append((discount_factor**j)*reward_list[i])
      j+=1
    Cummulative_reward_list_timestep = np.array(Cummulative_reward_list_timestep)
    return np.flip(Cummulative_reward_list_timestep)

  
  def incomeEvaluation(self,Action, day):
    #'BUY'  0
    #'HOLD' 1
    #'SELL' 2
    # Checking the income Generated for past user given days
    Stock_price_open = self.DC.data['Open'].iloc[day]
    Stock_price_close = self.DC.data['Close'].iloc[day]
    no_of_times_holded = 0
    reward = 0
    max_holds = 2
    
    
    if Action != 1:            # Checking for the hold condition first

      if Action ==0:           # Buying the stocks 
        if self.BUY == 0:      # Already Bought so holding the stocks
          if self.holds <= max_holds:
            self.holds +=1
            self.HOLD = 1
            self.BUY  = 0
            self.SELL = 0
            #print("BUY HOLD")
          else:
            self.income = Stock_price_close * self.stocks 
            self.BUY   = 1
            self.SELL  = 0
            self.HOLD  = 0
            self.holds = 0
            self.track_income.append(self.income)
            #print("BUY HOLD SOLD")
          return self.Rewards_metric[1]
        else:                  # Buying the stocks
          self.stocks = self.income // Stock_price_open
          self.income = self.income % Stock_price_open
          self.SELL = 1        # After Buying Customer Eagers to sold the stock
          self.BUY  = 0        # Putting the condition not to BUY stocks.
          self.HOLD = 0
          #print("BUY")
          return self.Rewards_metric[0]

      if Action == 2:          # Selling the Stocks.
        if self.BUY == 1:
          self.stocks = self.income // Stock_price_open
          self.income = self.income % Stock_price_open
          self.SELL = 1        # After Buying Customer Eagers to sold the stock
          self.BUY  = 0        # Putting the condition not to BUY stocks.
          self.HOLD = 0
          #print("SELL BUY")
          return self.Rewards_metric[0]
        else:
          if self.HOLD == 0:   # The selling stocks are not in hold
            self.income = Stock_price_close * self.stocks
            self.stocks = 0
            self.HOLD = 0
            self.BUY  = 1
            self.SELL = 0
            self.track_income.append(self.income)
            #print("STOCK SELL")
            return self.Rewards_metric[2]

          if self.HOLD == 1:  # The selling stocks are in hold and deducting the taxes.
            self.income = Stock_price_close * self.stocks
            self.stocks = 0
            self.HOLD = 0
            self.BUY  = 1
            self.SELL = 0
            self.track_income.append(self.income)
            #print("HOLD STOCK SELL")
            return self.Rewards_metric[2]
    else:
      # Doing the Holding condition
      # Deducting the taxes while selling in case of the hold by 5%
      if self.BUY == 1:
        self.stocks = self.income // Stock_price_open
        self.income = self.income % Stock_price_open
        self.SELL = 1        # After Buying Customer Eagers to sold the stock
        self.BUY  = 0        # Putting the condition not to BUY stocks.
        self.HOLD = 0
        #print("HOLD BUY")
        return self.Rewards_metric[0]
      else:
        if self.holds <= max_holds:
          self.holds +=1
          self.HOLD = 1
          self.BUY  = 0
          self.SELL = 1
          #print("HOLD")
        else:
          self.income = Stock_price_close * self.stocks 
          self.BUY = 1
          self.SELL = 0
          self.HOLD = 0
          self.holds = 0
          self.track_income.append(self.income)
          #print("HOLD SELL")
        return self.Rewards_metric[1]
  
  # Fucntion 3 => This function train the Networks with some data from the Observation 

  def Pre_experience_Actor_critic(self,X_train,Y_train):   #### Function 3
    # Training the Data to the Policy Network
    print("Training the Policy Network")
    Policy_trained = self.Policy_model.fit(X_train, Y_train, epochs=1, batch_size=1)
  
    print("Training the Critic Network")
    # Training the Data to the Critic Network
    Critic_trained = self.Critic_model.fit(X_train,Y_train, epochs =1, batch_size=1)
  
  # Function 3 =>

  def Actor_critic(self, X_train, Y_train):
    Predicted_Policy_Actions = self.Policy_model.predict(X_train)
    #print("Done Policy")

    # Predicting the Present Value State
    Present_Value_State = self.Critic_model.predict(X_train)
 
    # Fomating the shape of the Present Value State
    Present_Value_State = [i[0] for i in Present_Value_State]

    # Formation of the Next Value State
    Next_Value_State = []
    for i in range(len(Present_Value_State)):
      try:
        Next_Value_State.append(Present_Value_State[i+1][0])
      except:
        Next_Value_State.append(0)

    # Predicted Actions
    Predicted_actions = [np.argmax(i) for i in Predicted_Policy_Actions]

    # Rewards According to Action
    reward = [self.Rewards_metric[i] for i in Predicted_actions]

    # Finding the TD Error
    TD_Error = reward + ((self.gamma_value * np.array(Next_Value_State)) - np.array(Present_Value_State))

    returns = self.discounted_rewards(reward)

    Advantage = returns - Present_Value_State
    Advantage = torch.Tensor(Advantage)
  
    # Adding the Advantage for the Training data of the Actor Network
    X_train_Actor = []
    for i,j in zip(X_train,Advantage):
      i = np.append(i,j)
      X_train_Actor.append(i)
    X_train_Actor = np.array(X_train_Actor)

    # Training the Actor Network
    actor  = self.Actor_model.fit(X_train_Actor, Y_train, batch_size = 1)

    # Updating the weights of the policy network with the actor network
    modelWeights       = self.Actor_model.trainable_weights[2]
    targetModelWeights = self.Policy_model.trainable_weights[2]

    targetModelWeights.assign(modelWeights)

    modelWeights_bias       = self.Actor_model.trainable_weights[3]
    targetModelWeights_bias = self.Policy_model.trainable_weights[3]

    targetModelWeights_bias.assign(modelWeights_bias)

    # Training the Critic and Policy with the present Data 
    policy = self.Policy_model.fit(X_train, Y_train, batch_size = 1)
    critic = self.Critic_model.fit(X_train, Y_train, batch_size = 1)
    # Getting the Loss for every Network
    Actor_fit_loss  = np.mean(actor.history['loss'][0])
    policy_fit_loss = np.mean(policy.history['loss'][0])
    critic_fit_loss = np.mean(critic.history['loss'][0])

    return (Actor_fit_loss, policy_fit_loss, critic_fit_loss)
  
  def Implementing_the_actor_critic(self):
    # Training the Networks to have the prior knowledge on the data
    self.Pre_experience_Actor_critic(self.X_train[0:100],self.Y_train[0:100])
    # Traing the Actor Critic on the data
    Start_data = self.Start_data
    End_data   = self.End_data
    self.rewards = np.array([])
    while End_data <= len(self.X_train):
      self.income = 100
      print("Working on the Data From:", Start_data," to :",End_data)
      (Actor_fit_loss, policy_fit_loss, critic_fit_loss) = self.Actor_critic(self.X_train[Start_data:End_data],self.Y_train[Start_data:End_data])
      self.Actor.append(Actor_fit_loss)
      self.Critic.append(policy_fit_loss)
      self.Policy.append(critic_fit_loss)
      Start_data += self.data_size
      End_data   += self.data_size
      if Start_data % 100 ==0:
        output.clear()

  def test(self):
    predicted_probs = self.Policy_model.predict(self.X_test)
    predicted = np.array([])
    for i in predicted_probs:
      predicted = np.append(predicted, np.argmax(i))

    predicted_action = np.argmax(predicted_probs, axis =0)
    self.accuracy_score = accuracy_score(self.Y_test,predicted)
    print("Training Accuracy: ",self.accuracy_score*100)
  
  def income_returns(self):
    Action_probabilities = self.Policy_model.predict(self.DC.Observation_data_frame.iloc[-object_AC.no_of_days:-1:1,])
    predicted = np.argmax(Action_probabilities,axis = 1)

    true_value = self.Target[-self.no_of_days:-1:1]
    true_ = np.array([])
    for i in true_value:
      if i == 'SELL':
        true_ = np.append(true_,2)
      if i == 'BUY':
        true_ = np.append(true_,0)
      if i == 'HOLD':
        true_ = np.append(true_,1)
    return accuracy_score(true_,predicted)*100

  def Render(self):
    plt.figure(figsize=(10,8))
    plt.title('The Loss Graph of All three Networks')
    plt.plot(range(0,len(self.Actor)),self.Actor, label = "Actor Loss")
    plt.plot(range(0,len(self.Critic)),self.Critic, label = "Critic Loss")
    plt.plot(range(0,len(self.Policy)),self.Policy, Label = "Policy Loss")
    plt.xlabel('Training')
    plt.ylabel('Loss')
    plt.legend()
    plt.show() 

    
    plt.title('The Rewards Graph')
    plt.plot(range(0,len(self.rewards)),self.rewards, label = "Policy Loss")
    plt.xlabel('Training')
    plt.ylabel('Rewards')
    plt.legend()
    plt.show()

  def Render_income(self):
    plt.title('The Income Graph')
    plt.plot(range(0,len(self.track_income)),self.track_income, label = "Income Graph")
    plt.xlabel('Training')
    plt.ylabel('Income')
    plt.legend()
    plt.show()
    


object_AC = ACTOR_CRITIC(Object_DC)
```
